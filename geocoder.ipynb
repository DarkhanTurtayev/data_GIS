```python
import pandas as pd
import numpy 
import xdrlib
import xml
from pandas import DataFrame
import xlrd


```

```python
excel_data = pd.read_excel('/Users/macbookm1/Desktop/krisha mapping test new.xlsx')
df = pd.DataFrame(excel_data, columns=['Улица', 'Цена за квадрат'])
adresses = df['Улица'].tolist()
print(adresses[0:10])
```

```python
geo_data = []
lat_data = []
lon_data = []
for i in adresses:
    try:
         coordinates = []
         from geopy.geocoders import Nominatim
         geolocator = Nominatim(user_agent="TryTestApp")
         location = geolocator.geocode('Алматы' + i, language='ru')
         print(location.address)
         lat = location.latitude
         lon = location.longitude
         print((location.latitude, location.longitude))
         coordinates.append(lat)
         coordinates.append(lon)
    except:
        coordinates = ''
        lat = ''
        lon = ''
    geo_data.append(coordinates)
    lat_data.append(lat)
    lon_data.append(lon)
print(geo_data)
```

```python
dt = DataFrame({
    
    'Адрес' : adresses,
    'Цена' : df['Цена за квадрат'].tolist(),
    'Lat' : lat_data,
    'Lon' : lon_data,
    'Координаты' : geo_data

})

dt.to_excel('data')
```
