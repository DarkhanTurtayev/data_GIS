```python
excel_data = pd.read_excel('/Users/macbookm1/Desktop/krisha mapping test new.xlsx')
df = pd.DataFrame(excel_data, columns=['Улица', 'Цена за квадрат'])
adresses = df['Улица'].tolist()
prices = df['Цена за квадрат'].tolist()

new_file = dict(zip(adresses, prices))

```

```python
geo_data = []
lat_data = []
lon_data = []
for i in adresses:
    try:   
         coordinates = []
         from geopy.geocoders import Nominatim
         geolocator = Nominatim(user_agent="TryTestApp")
         location = geolocator.geocode('Алматы' + i, language='ru')
         print(location.address)
         lat = location.latitude
         lon = location.longitude
         print((location.latitude, location.longitude))
         coordinates.append(lat)
         coordinates.append(lon)
    except:
        coordinates = ''
        lat = ''
        lon = ''
    geo_data.append(coordinates)
    lat_data.append(lat)
    lon_data.append(lon)

print(geo_data)

dm = DataFrame({
    'Цена': prices,
    'Lat' : lat_data,
    'Lon' : lon_data,
    'Координаты' : geo_data,
    'Адрес' : adresses
 })

dm.to_excel('coordinates_test_2.xlsx')

```

```python
data = pd.read_excel('/Users/macbookm1/Desktop/Parser Prices/Mapper/coordinates_test_2.xlsx')
# Convert to dataframe and specific fields
df = pd.DataFrame(data, columns=['Lat', 'Lon', 'Цена'])
# Return a list
df.dropna(subset=['Lat'], inplace=True)
df

max_value = max(df['Цена'])
mean = sum(df['Цена'])/len(df['Цена'])
cof = []
for x in df['Цена']:

    calc = 1/mean * x 
    cof.append(calc)



dataset = list(map(list, zip(df["Lat"],
                          df["Lon"],
                          cof
                         )
               )
           )

print(dataset)
```

```python
from folium import *

heatmap= folium.Map(location=[43.238805, 76.945259],
                    zoom_start = 15)

HeatMap(dataset, blur= 10, gradient={0.2: 'purple', 0.3: 'blue', 0.5: 'yellow', 0.7: 'orange', 1.0: 'red'}, 
                min_opacity=0.3, 
                max_opacity= 1.0, 
                radius=30,
                use_local_extrema=False).add_to(heatmap)

heatmap

```
